IMPORTING PACKAGES

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

INITIALIZING

```python
timesteps = 30
starting_price = 600
volatility = 2/100
max_simulations = 5000
```

MCS CORE

```python
# creating a array
paths = np.full((timesteps+1 , max_simulations) , starting_price) 
paths[0][:] = starting_price
# tracking possible path and storing it in 2-d space created
for j in range (0 , max_simulations):
    for i in range (1 , timesteps+1):
        low = paths[i-1][j]*(1-volatility)
        high = paths[i-1][j]*(1+volatility)
        paths[i][j] = np.random.uniform(low , high)

final_price = paths[timesteps][:]
```

PLOTTING GRAPH

```python
plt.figure(1)
plt.plot(paths)
plt.xlabel("Time")
plt.ylabel("Stock Prices")
plt.figure(2)
plt.hist(final_price , bins=60)
plt.xlabel("Price")
plt.ylabel("No. of Occurence")

plt.show()
```
